```python
import pandas as pd

data = pd.read_csv('data.csv', index_col=0)
data
```

```python
# Mencari nilai mean dari data ukuran ruang kantor, tarif internet, dan harga sewa per bulan 
mean_ukuran_ruang_kantor = data['Ukuran'].mean()
mean_tarif_internet = data['Tarif Internet'].mean()
mean_harga_sewa_per_bulan = data['Harga Sewa'].mean()

resultMean = {
    'Ukuran Ruang Kantor': mean_ukuran_ruang_kantor,
    'Tarif Internet': mean_tarif_internet,
    'Harga Sewa Per Bulan': mean_harga_sewa_per_bulan
}

pd.DataFrame(resultMean, index=['Nilai Mean'])
```

```python
# Mencari nilai median dari data ukuran ruang kantor, tarif internet, dan harga sewa per bulan 
median_ukuran_ruang_kantor = data['Ukuran'].median()
median_tarif_internet = data['Tarif Internet'].median()
median_harga_sewa_per_bulan = data['Harga Sewa'].median()

resultMedian = {
  'Ukuran Ruang Kantor': median_ukuran_ruang_kantor,
  'Tarif Internet': median_tarif_internet,
  'Harga Sewa Per Bulan': median_harga_sewa_per_bulan
}

pd.DataFrame(resultMedian, index=['Nilai Median'])
```

```python

# Mencari nilai modus dari data ukuran ruang kantor, tarif internet, dan harga sewa per bulan
modus_ukuran_ruang_kantor = data['Ukuran'].mode()
modus_tarif_internet = data['Tarif Internet'].mode()
modus_harga_sewa_per_bulan = data['Harga Sewa'].mode()

resultModus = {
  'Ukuran Ruang Kantor': modus_ukuran_ruang_kantor,
  'Tarif Internet': modus_tarif_internet,
  'Harga Sewa Per Bulan': modus_harga_sewa_per_bulan
}

pd.DataFrame(resultModus, index=['Nilai Modus'])
```
